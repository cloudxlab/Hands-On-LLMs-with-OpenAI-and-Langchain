```python
import numpy as np
X = np.array([1, 2, 3])
y = np.array([2, 4, 3.5])


def error(m, c, X, y):
    y_pred = m*X + c
    diff = y_pred - y
    dsq = diff *diff
    return np.sum(dsq)

# Adam
# Nasterov + ADAM
```

```python
error(1, 0, X, y)
```

```python
m = 5
c = 6
delta = 1e-7
def observe_m():
    prev_e = error(m, c, X, y)
    new_e = error(m+delta, c, X, y)
    return (new_e-prev_e)/delta

def observe_c():
    prev_e = error(m, c, X, y)
    new_e = error(m, c+delta, X, y)
    return (new_e-prev_e)/delta
```

```python
grad_m = observe_m()
if grad_m > 0:
    m = m - grad_m * 0.01
else:
    m = m + -grad_m * 0.01
print(m)
```

```python
for i in range(200):
    print("-----")
    print("Epoch: ", i)
    grad_m = observe_m()
    grad_c = observe_c()
    print("Gradient: ", grad_m)
    m -= grad_m * 0.02
    c -= grad_c * 0.02
    print("New Error: ", error(m, c, X, y))
    print(m, c)
```

```python
def sq(x):
    return x*x
```

```python
x = 10
y = sq(x)
print("y: {y}")

x1 = x + delta
y1 = sq(x1)

# dy/dx - rate of change of y with respect to x
print("dy/dx: ", (y1-y)/delta)
```
