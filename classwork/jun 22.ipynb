```python
# Sushanta 
from random import random
import numpy as np
def weighted_probs(probs):
    enum_list = list(enumerate(probs))
    r = random()
    probability = 0
    for i in enum_list[:-1]:
        probability +=i[1]
        if r<probability:
            return i[0]
    return enum_list[-1][0]
probs  = [0.2,0.3,0.4,0.1]
s = [weighted_probs(probs) for _ in range(1000)]
s = np.array(s)
len(s[s==0])/len(s),len(s[s==1])/len(s),len(s[s==2])/len(s),len(s[s==3])/len(s)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def display_proba(arr):
    # Generate x-axis indices (assuming each bar represents a category)
    x = range(len(arr))

    # Plotting the bar chart
    plt.bar(x, arr, color='blue')

    # Adding labels and title
    plt.xlabel('Words')
    plt.ylabel('Probabilities')
    plt.title('Bar Chart of Heights')

    # Adding x-axis labels with category names if needed
    # Replace ['Category1', 'Category2', ...] with actual category labels
    plt.xticks(x, ['Play', 'Fight', 'Live'])

    # Display the plot
    plt.show()

# Example array of heights (you can replace this with your own data)
arr = [0.45, 0.40, .15]
display_proba(arr)
```
